# X,O

```python
from IPython.display import clear_output

def display_board(board):
    clear_output()
    print (" "+board[0]+"|"+board[1]+"|"+board[2])
    print("---------")
    print (" "+board[3]+"|"+board[4]+"|"+board[5])
    print("---------")
    print (" "+board[6]+"|"+board[7]+"|"+board[8])
def player_input():
    marker=""
    while True:
        marker =input("player 1: do you want to be x or o?").lower()
        if(marker!="x"and marker!= "o"):
            print("invailed input")
        elif marker=="x":
            return ["x","o"]
        else:
            return ["o","x"]
        
def place_marker(board, marker, position): # example: place_marker(board, 'X', 4)
    
    board[position-1]=marker

def win_check(board, marker):
    
    return ((board[0]==marker and board[1]==marker and board[2]==marker) or
            (board[3]==marker and board[4]==marker and board[5]==marker) or
            (board[6]==marker and board[7]==marker and board[8]==marker) or
            (board[0]==marker and board[3]==marker and board[6]==marker) or
            (board[1]==marker and board[4]==marker and board[7]==marker) or
            (board[2]==marker and board[5]==marker and board[8]==marker) or
            (board[0]==marker and board[4]==marker and board[8]==marker) or
            (board[2]==marker and board[4]==marker and board[6]==marker))
import random

def choose_first_v1():
    player=random.randint(1,2)
    if player==1:
        return ("player1")
    else :
        return ("player2")
    
def space_check(board, position):
    return (board[position-1]==" ")
def full_board_check(board):
    return(" "not in board)
def player_choice(board):
    position=0
    while True:
        try:
            position=int(input("please choose your next position from available positions(1,9)!"))
        except:
            print("enter a number")
        else:
            if position<1 or position>9:
                print ("invaild position")
            elif(not space_check(board,position)):
                print("this position is already filled ,choose another position")
            else:
                return position
def replay():
    while True:
        cun=input("do you want to continuo?\n(y,n)\n")
        if (cun.lower() in ["y","n"]):
            break
        
    if(cun.lower()!="y"):
        return False
    else:
        return True
print('Welcome to Tic Tac Toe!')
import time
while True:
    board=[' ',' ',' ',' ',' ',' ',' ',' ',' ']
    vals=player_input()
    marker1=vals[0]
    marker2=vals[1]
    turn=(choose_first_v1())
    print(turn+"will go first")
    game_on=True
    time.sleep(3)
    while game_on:
        display_board(board)
        if turn=="player1":
            board[player_choice(board)-1]=marker1
            if win_check(board,marker1):
                display_board(board)
                print("player 1 won the game")
                game_on=False
            elif full_board_check(board):
                print("it's a draw!")
                game_on=False
            else:
                turn="player2"
        else:
            position=player_choice(board)
            place_marker(board,marker2,position)
            if win_check(board,marker2):
                display_board(board)
                print("player 2 won the game")
                game_on=False
            elif full_board_check(board):
                print("it's a draw!")
                game_on=False
            else:
                turn="player1"
    if not replay():
        break
```
